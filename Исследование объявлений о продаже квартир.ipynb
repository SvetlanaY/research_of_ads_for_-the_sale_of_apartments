####  заполнение пропусков по столбцу ceiling_height

```python
# посмотрим статистические данные по ceiling_height
df['ceiling_height'].describe()
```

Посмотрим зависимость цены на квартру от высоты потолка. Предполагаем, что квартиры с нестандартно высоким потолком должны быть дорогими.

```python
# для понижения масштаба по столбцу last_price, прологарифмируем данный столбец
df['last_price_log']=np.log(df['last_price'])
```

```python
# Посмотрим зависимость цены на квартру от высоты потолка
df.plot(x='ceiling_height',
        y='last_price_log',
        kind='scatter',
        grid='True',
        figsize=(15,4),
        title='Зависимость цены на квартиру от высоты потолка')
plt.show()
```

1. При значении ceiling_height более 10 метров цена на квартиры не растет, и даже немного падает. Скорее всего высота потолков указанная более 10 метров это ошибка, приблизим участок графика в интервале до 10 метров. 
2. **В интервале от 20 до 40 достаточно большое количество квартир, можно предположить, что высота потолка указана не в метрах, а дециметрах,т.е. реальная высота потолков от 2 до 4 метров. Необходимо заменить эти данные в датафрейме.**

```python
# Посмотрим зависимость цены на квартру от высоты потолка при высоте потолка менее 10 метров
(df.query('ceiling_height <= 10')
   .plot(x='ceiling_height',
         y='last_price_log',
         kind='scatter',
         grid='True',
         figsize=(15,4),
         title='Зависимость цены на квартиру от высоты потолка')
)
plt.show()
```

При значении высоты потолка до 5 метров достаточно высокая цена, но при 8 метрах цена уже ниже.Между 5-6 метрами единичные квартиры с высоким разбросом по стоимости квартиры. Исходя из этого предполагаем, что высота потолка 5 и более метров указана ошибочно, необходимо посмотреть данные и заполнить более подходящими значениями.

```python
print('Количество квартир с высотой потолка более или равно 5 метров, не входящих в интервал от 20 до 40 метров:',
      df[(df['ceiling_height'] >= 5) & 
         ~((df['ceiling_height'] > 20) & (df['ceiling_height'] < 40))].shape[0])
```

**По данным 16 квартирам необходимо заменить данные на те же значения, которыми будем заполнять пропущенные ячейки**

```python
# Посчитаем количество квартир с высотой потолка менее 2 метра.
len(df.query('ceiling_height < 2'))
```

**По квартирам с высотой потолка менее 2 метра (3 квартиры) необходимо заменить данные на теже значения, которыми будем заполнять пропущенные ячейки.**

```python
# Посмотрим гистограмму распределения высоты потолка для данных менее 6 метров и более или равно 2,2 метра
df['ceiling_height'].hist(range=(2,5),figsize=(15,4),bins=50)
plt.xlabel('Высота потолка, м.')
plt.ylabel('Частота')
plt.title("Гистограмма распределения высоты потолка")
plt.show()
```

Гистограмма распределения ожидаема, самые распространенные квартиры с высотой потолка примерно от 2,5 метра до 2.8 метров.  Для заполнения пропущенных значений возьмем среднее значение из интервала от 2 до 5 метров.

```python
# рассчитаем среднее значение из интервала от 2 до 5 метров.
ceiling_height_for_nan = round(df.query('ceiling_height >= 2 and ceiling_height < 5')['ceiling_height'].mean(),2)
ceiling_height_for_nan
```

```python
# Заполним пропущенные данные средним значением, рассчитанным выше
df['ceiling_height'].fillna(ceiling_height_for_nan, inplace=True)
```

```python
# проверим, что больше не осталось пустых значений по столбцу ceiling_height
df['ceiling_height'].isna().sum()
```

```python
# В интервале от 20 до 40 заменим данные в датафрейме, разделив исходные на 10
df.loc[(df['ceiling_height'] > 20) & (df['ceiling_height'] < 40),'ceiling_height'] = df.loc[(df['ceiling_height'] > 20) & (df['ceiling_height'] < 40),'ceiling_height'] / 10
```

```python
# По квартирам с высотой потолка менее 2 метра заменим данные на ceiling_height_for_nan
df.loc[df['ceiling_height'] < 2,'ceiling_height'] = ceiling_height_for_nan
```

```python
# По квартир с высотой потолка более 5 метров заменим данные на ceiling_height_for_nan
df.loc[df['ceiling_height'] >= 5,'ceiling_height'] = ceiling_height_for_nan
```

```python
# проверим статистику по ceiling_height
df['ceiling_height'].describe()
```

**Вывод по столбцу ceiling_height**: 
- пропущенные значения заменили на средние , при расчете среднего не учитывались квартиры с высотой потолка более или равно 5 метров и менее 2 метра  
- значения в интервале 20-40 метров заменены на значения, разделенные на 10  
- значения в интервале меньше 2 и больше или равно 5 метров заменены на среднее, поссчитанное для заполнения пустых ячеек  

Ограничение по высоте в 5 метров получили по графику взаимосвязи стоимости квартиры и высоты потолка.
